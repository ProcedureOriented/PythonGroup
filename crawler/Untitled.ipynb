```python
import os
import re
class blg:
    def __init__(self,mon,day,con,nam,atc,coc,rec):#月，日，内容，博主，点赞，评论，转发
        self.mon=mon
        self.day=day
        self.con=re.sub(r'<(.*?)>','',con)#去除多余内容
        self.con=self.con.replace("content：",'')
        self.nam=nam
        self.atc=atc
        self.coc=coc
        self.rec=rec
def cmpa(a,b):#自定义匹配函数
    for i in b:
        if a.find(i)!=-1:
            return True
    return False
lis=[]
with open('data.txt','r')as f:
    while True:
        s=f.readline()
        t=re.match(r'name:(.*)',s)#匹配博主
        if s=='0' or s=='0\n' or s=='':
            break
        if t:#匹配到具体博主后，开始匹配博客
            na=s.replace('name:','')
            #print(na)
            q=f.readline()
            while True:
                p=f.readline()
                if p=='0'or p=='0\n' or p=='' or p=='-1\n':
                    break
                if re.match(r'URL：',p):#博客以第一行URL为起始，后接第二行时间，第三行内容，第四五六行赞/评论/转发
                    t=f.readline()
                    t=t.replace('created at：','')
                    tls=t.split('-')
                    if len(tls)!=2:#有两种情况会导致拆分出的不是月日两个。一是近一天内的信息，二是其他年份的信息，都不需要考虑
                        continue
                    mm=int(tls[0])
                    dd=int(tls[1])#分解为月、日
                    cont=f.readline()
                    atco=f.readline()
                    coco=f.readline()
                    reco=f.readline()
                    aa=int(atco.replace('attitudes count：',''))
                    cc=int(coco.replace('comments count：',''))
                    rr=int(reco.replace('reposts count：',''))
                    bg=blg(mm,dd,cont,na,aa,cc,rr)
                    lis.append(bg)
                    print(bg.nam)
        t=None
ls=[]
for i in range(1,6):
    for j in range(1,32):
        for k in lis:
            if k.mon==i:
                if k.day==j:
                    ls.append(k)#将博主为主的排列变为时间为主
with open('list.txt')as f:
    wlst=f.readlines()
n=len(ls)
with open('result.txt','a')as f:
    for i in wlst:
        f.write('关键词: %s'%i)
        j=0
        t=i.replace('\n','')
        b=t.split()
        while j<n:
            if cmpa(ls[j].con,b):#查到弱某一日有，则就地开始对本日的统计
                m=ls[j].mon
                d=ls[j].day
                f.write('日期：%d-%d\n'%(m,d))
                t1,t2,t3,t4=0,0,0,0
                h=j
                while h<n and ls[h].mon==m and ls[h].day==d :
                    if cmpa(ls[h].con,b):
                        t1+=1
                        t2+=ls[h].atc
                        t3+=ls[h].coc
                        t4+=ls[h].rec
                        f.write('内容：%s博主：%s 点赞数：%d，评论数：%d，转发数：%d\n\n'%(ls[h].con,ls[h].nam,ls[h].atc,ls[h].coc,ls[h].rec))
                    h+=1
                f.write('本日共博客数：%d，点赞数：%d，评论数：%d，转发数：%d\n\n'%(t1,t2,t3,t4))
                j=h
            else:
                j=j+1
print ("complete")
```
